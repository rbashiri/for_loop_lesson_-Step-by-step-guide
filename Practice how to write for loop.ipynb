## A for loop means:

        “Take one item at a time from a group, and do something with it.”
**for item in group:**
    do_something
### Step 1: Start with your data
- You need something that contains multiple items.

        numbers = [3, 7, 10, 15] This is your group.
### Step 2: Decide what name you want for one item
- Inside the loop, Python will take one value at a time from the list.
- You choose a temporary name for that one value.
        for num in numbers:
        
         - Here:
         - numbers = the whole list
         - num = one item from that list
         - So Python reads this like:
         - take 3, call it num
         - take 7, call it num
         - take 10, call it num
         - take 15, call it num
### Step 3: Add a colon
    A for loop line must end with :
    for num in numbers:
    That colon means:
    “The loop body starts now.”

### Step 4: Indent the next line

        Everything inside the loop must be indented.
### Universal template

        This is the structure you can apply to many problems:

group = [...]

        for item in group:
        # do something with item
### How to think before writing the loop

#### Ask yourself these 3 questions:

        1. What is my group?

                Example:
                list of numbers
                list of names
                string
                rows in data

        2. What is one item?

                Example:
                one number
                one student name
                one character

        3. What do I want to do with each item?

                Example:
                print it
                count it
                add it
        check a condition
        save it into a new list    

```python
numbers = [3, 7, 10, 15]
for num in numbers:
    print(num)
```

```python
fruits = ["apple", "banana", "kiwi"]
print(fruits)
for furit in fruits:
    print(furit)
    
```

```python
# Example 2: Count something
# Problem: count how many numbers are greater than 10. There is conditional logic involved, so we need to use an if statement. We also need to keep track of the count, so we need a variable to store that count. We will use a for loop to iterate through the list of numbers.

numbers = [3,12,7,20,15]
# start conter at 0
count = 0
for num in numbers: # loop through each number in the list
    if num > 10: # check if the number is greater than 10
        count  += 1 # if it is, increment the counter by 1
print(count)
```

```python
# Example 3: Build a new list
# Problem: keep only numbers greater than 10.

numbers = [3,12,7,20,15]
# create an empty list to store the numbers greater than 10
greater_than_10 = [] # Create an empty list to store answers.

for num in numbers:
    if num >10:
        greater_than_10.append(num) # if the number is greater than 10, add it to the new list
print(greater_than_10)
```

```python
# Example 4: Sum values
# Problem: calculate total of numbers in a list.
numbers = [3,12,7,20,]

total = 0 # You need a place to store the running sum. create a variable to store the total, starting at 0
for num in numbers: # Take one number at a time.
    total += num # Keep adding each number.
    #This becomes: total = 0 + 3 = 3, 
    # total = 3 + 12 = 15
    # total = 15 + 7 = 22
    # total = 22 + 20 = 42
print(total)
```

## The 4 reusable for loop patterns

## These are the ones you should memorize.

    1. Print pattern
    for item in group:
     print(item)

Use when you just want to see each value.

    2. Count pattern
    count = 0

    for item in group:
        if condition:
            count += 1

## Use when you want:
how many
count of values
number of matches

    3. Sum pattern
    total = 0

    for item in group:
        total += item

## Use when you want:
total
average
running sum

    4. Filter pattern
    result = []

    for item in group:
        if condition:
            result.append(item)

## Use when you want:
keep only some items
build a new list
clean data
One master template for many applications

    print(result)

        How to write a for loop from scratch every time

## Use this exact thinking process.

Case A: I want to print each item

    1- Write the data
    2- Write for
    3- Choose a temporary item name
    4- Write in
    5- Write the group name
    6- Add :
    7- Indent
    8- Write what to do

Example:
    students = ["Ali", "Sara", "John"]
    for student in students:
        print(student)

## Case B: I want to count items

- Write the data
- Make counter = 0
- Start loop
- Check condition
- Increase counter
= Print final counter

Example:

    ages = [15, 22, 18, 30, 12]
    count = 0

    for age in ages:
        if age >= 18:
            count += 1
print(count)

## Case C: I want to make a new list

- Write the data
- Create empty list
- Start loop
- Check condition
- Append matching item
- Print new list

    Example:
    ages = [15, 22, 18, 30, 12]
    adults = []
    for age in ages:
        if age >= 18:
            adults.append(age)
    print(adults)

**Common mistakes**

* Mistake 1: Forgetting colon
Wrong:
for num in numbers
* Correct:

for num in numbers:

* Mistake 2: No indentation
Wrong:
    for num in numbers:
    print(num)

* Correct:

    for num in numbers:
        print(num)

* Mistake 3: Using the list name instead of item name

    Wrong:

    numbers = [1, 2, 3]

    for num in numbers:
        print(numbers)

* This prints the whole list each time.

* Correct:
    for num in numbers:
        print(num)

* Mistake 4: Forgetting to initialize counter or result

    Wrong:
    for num in numbers:
    count += 1

*Correct:

    count = 0
    for num in numbers:
        count += 1
